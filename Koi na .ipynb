```
Write a “pager” program. Your solution should prompt for a filename, and display the text file 25 lines at a time, pausing
each time to ask the user to enter the word “continue”, in order to show the next 25 lines or enter the word “stop” to
close the file.

````python
def pager():
    with open("alice1.txt",encoding="UTF-8") as f:
        lines = f.readlines()
        totallines = len(lines)
        start = 0
        end = min(25,totallines)
        while start<end:
            for i in range(start,end):
                print(i,lines[i],end="")
            user = input("Enter continue to continue and stop to stop")
            if user == 'stop':
                break
            start = start + 25
            end = min(start+25,totallines)
````

````python
pager()
````

Write a python program to read a text file “Story.txt” and print only word starting with ‘I’ in reverse order.
Example: If value in text file is : ‘INDIA IS MY COUNTRY’
Output will be: ‘AIDNI SI MY COUNTRY

````python
%%writefile Story11.txt
INDIA IS MY COUNTRY
````

````python
result = []
with open("Story11.txt") as f:
    for line in f:
        for word in line.split():
            if word[0]=='i' or word[0]=='I':
                result.append(word[::-1])
            else:
                result.append(word)
        print(" ".join(result))
````

````python
with open("comment11.txt") as f, open("comment22.txt", "w") as f1:
    for line in f:
        line = line.strip()

        # Skip full comment lines
        if line.startswith("#"):
            continue

        # Remove inline comments
        if "#" in line:
            line = line.split("#")[0]

        f1.write(line + "\n")
````

## WAP to count number of lines, words, characters including spaces in a given line
```
sample file -
this is python
this is not java

output:
lines = 2
words = 7
characters = 30
```

````python
%%writefile sample.txt
this is python
this is not java
````

````python
with open("sample.txt") as file:
    content = file.read()
    file.seek(0)
    linecount = len(file.readlines())
    wordcount = len(content.split())
    charcount = len(content) - content.count("\n")
    print("line = ",linecount)
    print("words = ",wordcount)
    print("charcount = ",charcount)
````

## WAP to display odd lines in a given file
```
input-
this is first line 
this is second line
this is third line 
this is forth line 
output-
this is first line

```

````python
f = open("odd.txt")
lineno = 1
for line in f:
    if lineno % 2 == 1:
        print(line, end="")
    lineno+=1
f.close()
    
````

# PB-474 Write a python program to search for a string in text files.

````python
def search(filename,query):
    with open(filename) as file:
        content = file.read()
        if query in content:
            print("The string is in the file")
        else:
            print("The string is not in the file")
search("friends.txt","honest")
````

## PB-475  Write a ‚Äúpager‚Äù program. Your solution should prompt for a filename, and display the text file 25 lines at a time, pausing each time to ask the user to enter the word ‚Äúcontinue‚Äù, in order to show the next 25 lines or enter the word ‚Äústop‚Äù to close the file. 

````python
def pager(filename):
    with open(filename, encoding="UTF-8") as file:
        lines = file.readlines()
        total_lines = len(lines)
        start = 0
        end = min(25,total_lines)
        while start<total_lines:
            for i in range(start,end):
                print(i,lines[i], end="")
            user = input("Enter text to continue or stop to stop")
            if user == 'stop':
                break
            start+=25
            end = min(start+25,total_lines)
pager("alice1.txt")
````

## PB-478 Write a python program to read a text file “Story.txt” and print only word starting with ‘I’ in reverse order.
Example: If value in text file is : ‘INDIA IS MY COUNTRY’
Output will be: ‘AIDNI SI MY COUNTRY’ 

````python
%%writefile Story.txt
INDIA IS MY COUNTRY 
````

````python
result = ""
with open("Story.txt") as f1:
    content = f1.read()
    s = content.split()
    for i in s:
        if i[0].lower() == 'i':
            result = result+i[::-1] + " "
        else:
            result = result+i + " "
    result = result.strip() + "\n"
print(result.strip("\n"))
    
````

### try except else finally
**SYNTAX**
```python
try:
    statement # this will run first
except:
    statement # this will run if error occurs
else:
    statement # this will run if no error occurs
finally:
    statement # this will run no matter what

```

````python
try:
    print("try something here ")
except:
    print("error occured")
else:
    print("Succesful run")
finally:
    print("this will run no matter what")
````

````python
try:
    print("try something here ")
    x=1/0
except:
    print("error occured")
else:
    print("Succesful run")
finally:
    print("this will run no matter what")
````

##  Creating custom error 

````python
class MyError(Exception):
    pass
class FiveDivisionError(MyError):
    pass
try:
    number = int(input("Enter a number - "))
    if number == 5:
        raise FiveDivisionError("Enter other number")
except FiveDivisionError:
    print("Enter number except 5")
````

````python
class MyError(Exception):
    pass
class FiveDivisionError(MyError):
    pass
try:
    number = int(input("Enter a number - "))
    if number == 5:
        raise FiveDivisionError("Enter other number")
finally:
    print("Done")
````

## CONSTRUCTOR METHOD
- This method is used to instantiate the object using `__init__`

````python
class Employee:
    employeecourt = 0
    def __init__(self,name,salary):
        print("Object instantiated")
        self.name = name
        print(id(self.name))
        self.salary=salary
        Employee.employeecourt+=1
    def display(self):
        return f"The name is {self.name} and salary is {self.salary}"
    def __str__(self):
        print("The object referance is destroyed and garbage collected")
    def __del__(self):
        print("The object referance is destroyed and garbage collected")
````

````python
emp1 = Employee('python',1000)
````

````python
emp1.name
````

````python
id(emp1.name)
````

````python
emp1.display()
````

````python
Employee.display()
````

````python
Employee.employeecourt
````

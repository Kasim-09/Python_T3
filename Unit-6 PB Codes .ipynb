# PB-475 
```
Write a “pager” program. Your solution should prompt for a filename, and display the text file 25 lines at a time, pausing
each time to ask the user to enter the word “continue”, in order to show the next 25 lines or enter the word “stop” to
close the file.
```

````python
def pager(filename):
    with open(filename , encoding="UTF-8") as file:
        lines = file.readlines()
        total_lines = len(lines)
        start = 0
        end = min(25,total_lines)
        while start<end:
            for i in range(start,end):
                print(i,lines[i], end="")
            user = input("Enter text to continue or stop to stop")
            if user == "stop":
                break
            start = start + 25
            end = min(start+25,total_lines)
````

````python
pager("alice1.txt")
````

# PB-479
```
Write a Python program to count words, characters and spaces from text file.
Input:
Python is a Easy Subject
OOPs is One of the most
interesting Topic
Output:
No of space: 10
No of word: 13
No of character: 64
```

````python
%%writefile interesting.txt
Python is a Easy Subject
OOPs is One of the most
interesting Topic
````

````python
countspace=0
charcount=0
with open("interesting.txt") as f:
    content = f.read()
    for i in content:
        if i == " ":
            countspace+=1
    s = content.split()
    wordcount = len(s)
    for i in s:
        charcount = len(i)+charcount
print(f"Space count = {countspace}")
print(f"Word count = {wordcount}")
print(f"Character count = {charcount}")

````
